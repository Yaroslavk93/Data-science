```python
Q1 = df['rectal temperature'].quantile(0.15)
Q3 = df['rectal temperature'].quantile(0.95)
iqr = Q3 - Q1
lower_bound = Q1 - (1.5 * iqr)
upper_bound = Q3 + (1.5 * iqr)
remove_outliers = df[df['rectal temperature'].between(lower_bound, upper_bound)]
remove_outliers
```

```python
df[~df['rectal temperature'].between(lower_bound, upper_bound, inclusive=True)]
```

```python
repl_val = df.copy()
```

```python
repl_val = repl_val.dropna(thresh=6)
repl_val
```

```python
repl_val[(repl_val['surgery?'] != 1) & (repl_val['surgery?'] != 2)]
```

```python
repl_val['surgery?'] = repl_val['surgery?'].fillna(repl_val['surgery?'].mode()[0])
repl_val[(repl_val['surgery?'] != 1) & (repl_val['surgery?'] != 2)]
```

```python
repl_val[repl_val['rectal temperature'].isnull()]
```

```python
# repl_val['rectal temperature'] = repl_val['rectal temperature'].fillna(repl_val['rectal temperature'].median())
repl_val['rectal temperature'] = repl_val['rectal temperature'].fillna(repl_val.groupby('surgery?')['rectal temperature'].transform('median'))
repl_val[repl_val['rectal temperature'].isnull()]
```

```python
repl_val[repl_val['pulse'].isnull()]
```

```python
repl_val['pulse'] = repl_val['pulse'].fillna(repl_val.groupby(['surgery?', 'rectal temperature'])['pulse'].transform('median'))
repl_val[repl_val['pulse'].isnull()]
```

Удаляем столбец с частотой дыхания, т.к. в документации указано, что норма 8 - 10. Полезность сомнительна при больших колебаниях 

```python
repl_val[repl_val['respiratory rate'] < 15]
```

```python
repl_val = repl_val.drop(['respiratory rate'], axis=1)
```

```python
(repl_val.isna().mean()*100).round(2)
```

```python
repl_val[repl_val['temperature of extremities'].isnull()]
```

```python
def temp_ext(table):
    if table['rectal temperature'] > 38.5:
        return 2
    elif table['pain'] == 4 or table['pain'] == 5:
        return 4
    return 1
```

```python
repl_val.loc[repl_val['temperature of extremities'].isnull(), 'temperature of extremities'] = repl_val.apply(temp_ext, axis=1)
repl_val[repl_val['temperature of extremities'].isnull()]
```

```python
(repl_val.isna().mean()*100).round(2)
```

```python
repl_val[repl_val['pain'].isnull()]
```

```python
def pain_val(table):
    if (table['temperature of extremities'] == 3) or (table['temperature of extremities'] == 4):
        return random.choice((4, 5))
    elif (table['temperature of extremities'] == 2) or (table['surgery?'] == 2):
        return random.choice((2, 3))
    return 1
```

```python
repl_val.loc[repl_val['pain'].isnull(), 'pain'] = repl_val.apply(pain_val, axis=1)
repl_val[repl_val['pain'].isnull()]
```

```python
repl_val['outcome'] = repl_val['outcome'].fillna(repl_val['outcome'].mode()[0])
```

```python
(repl_val.isna().mean()*100).round(2)
```
