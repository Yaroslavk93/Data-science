### Задание 1
Дан объект в $2D$ пространстве

```python
import numpy as np
import matplotlib.pyplot as plt

A = [
    [-100, -100, 1], 
    [-100, 100, 1], 
    [100, 100, 1], 
    [0, 0, 1], 
    [100, -100, 1], 
    [-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
```

При помощи линейных отображений:

- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x (поможет матрица `[[0, 1, 0], [1, 0, 0], [0, 0, 1]]`)

```python
plt.plot(x / 2, y / 2)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
```

```python
x2 = x * np.cos(130) + y * np.sin(130)
y2 = -x * np.sin(130) + y * np.cos(130)
```

```python
plt.plot(x2, y2)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
```
