### Задание 2
Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

- Найдём отдельно вероятность событий 1, 2 и 3 монет

```python
# Всего 2 исхода в каждом событии
m1 = 0.5
m2 = 0.5
m3 = 0.5
```

- Вероятность наступления событий А1 и А2

```python
# Всего 4 исхода в каждом событии
A1 = 0.25
A2 = 0.25
```

- Найдём вероятность наступления события A1_A2 (то есть A1 и A2 одновременно)

```python
# 16 исходов в событии A1_A2
A1_A2 = 1 / 16
```

- Если выполняется формула P ( A1_A2 ) = P ( A1 ) ⋅ P ( A2 ) , то события \(A1\) и \(A2\) являются независимыми.

```python
if A1_A2 == A1 * A2:
    print('События A1 и A2 являются независимыми.')
```

- в событии А1 ноходятся монеты 1 и 2;
- в событии А2 находятся монеты 2 и 3;
- соответсвенно событие А3 произошло, т.к. все 3 монеты уже были задействованы, и все 3 события в совокупности являются зависимыми

### Задание 3
Изучите 3 распределения из библиотеки scipy: [Нормальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), [Экспоненциальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html), [Стьюдента](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html). Используя функцию плотности распредления (`pdf`), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.  

Дополнительная информация [тут](https://pythonguides.com/scipy-stats/).

```python
import numpy as np
from scipy.stats import norm
from scipy.stats import expon
from scipy.stats import t
import matplotlib.pyplot as plt
```

Нормальное распределение

```python
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, norm.pdf(x), 'r-', lw=5, alpha=0.6, label='norm pdf')
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, norm.pdf(x, loc=0.5, scale=0.9), 'r-', lw=5, alpha=0.6, label='norm pdf')
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, norm.pdf(x, loc=2, scale=4), 'r-', lw=5, alpha=0.6, label='norm pdf')
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, norm.pdf(x, loc=10, scale=20), 'r-', lw=5, alpha=0.6, label='norm pdf')
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(x, norm.pdf(x, loc=30, scale=10), 'r-', lw=5, alpha=0.6, label='norm pdf')
```
